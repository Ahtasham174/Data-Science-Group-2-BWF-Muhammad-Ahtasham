### Given a normal distribution with a mean of 50 and a standard deviation of 5, find the probability that a value is between 45 and 55.

```python
from scipy import special,stats # used form finding norm and combination and other mathematical function
mean = 50
std_dev = 5
prob_between_45_55 = stats.norm.cdf(55, mean, std_dev) - stats.norm.cdf(45, mean, std_dev)
print(prob_between_45_55)
```

### Calculate the z-score for a value of 60 in a normal distribution with a mean of 50 and a standard deviation of 10.

```python
value = 60
mean = 50
std_dev = 10
z_score = (value - mean) / std_dev
print(z_score)
```

### If a coin is flipped 10 times, what is the probability of getting exactly 6 heads?

```python
n = 10
k = 6
p = 0.5
prob_6_heads = special.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
print(prob_6_heads)
```

### In a survey, 60% of people prefer chocolate ice cream. What is the probability that out of 8 randomly selected people, exactly 5 prefer chocolate ice cream?

```python
p_chocolate = 0.6
n_people = 8
k_chocolate = 5
prob_5_chocolate = special.comb(n_people, k_chocolate) * (p_chocolate ** k_chocolate) * ((1 - p_chocolate) ** (n_people - k_chocolate))
print(prob_5_chocolate)
```

### A call center receives an average of 5 calls per hour. What is the probability of receiving exactly 3 calls in an hour?

```python
lambda_calls = 5
k_calls = 3
prob_3_calls = (lambda_calls ** k_calls * np.exp(-lambda_calls)) / special.factorial(k_calls)
print(prob_3_calls)
```

### If a book has an average of 2 misprints per page, what is the probability that a page has no misprints?

```python
lambda_misprints = 2
k_misprints = 0
prob_no_misprints = (lambda_misprints ** k_misprints * np.exp(-lambda_misprints)) / special.factorial(k_misprints)
print(prob_no_misprints)
```
