# Task 16: Introduction to Matplotlib (basic plots: line, scatter, bar)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Part 1: Line Plot

### Creating a Simple Line Plot

```python
x = [0, 1, 2, 3, 4, 5]
y = [4, 9, 2, 10, 12, 8] 

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

### Plotting Multiple Lines on the Same Graph

```python
y1 = [6, 7, 5, 11, 9, 10] 
y2 = [8, 4, 6, 10, 13, 7] 
y3 = [17, 18, 19, 20, 9, 17] 
plt.plot(x, y1, label='Line 1', color='blue')
plt.plot(x, y2, label='Line 2', color='green')
plt.plot(x, y3, label='Line 2', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')
plt.legend()
plt.show()
```

### Adding Markers to a Line Plot

```python
x1 = np.linspace(1, 10, 30)
y4 = np.sin(x1)
y5 = np.cos(x1) 
y6 = np.log(x1)
y7 = np.log10(x1)

plt.plot(x, y, marker='o', color='red')
plt.plot(x, y1, marker='x', color='green')
plt.plot(x, y2, marker='.', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Markers("o", "x", ".")')
plt.show()

plt.plot(x1, y4, marker='^', color='brown')
plt.plot(x1, y5, marker='v', color='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Markers("^", "v")')
plt.grid(True) # for proper visualization
plt.show()

plt.plot(x1, y6, marker='<', color='maroon')
plt.plot(x1, y7, marker='>', color='orange')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Markers("<", ">")')
plt.grid(True) # for proper visualization
plt.show()
```

### Customizing Line Styles and Colors

```python
plt.plot(x, y, linestyle='-', color='purple', label='Solid line')
plt.plot(x, y1, linestyle='--', color='red', label='Dashed Line')
plt.plot(x, y2, linestyle='-.', color='green', label='Dash-dot Line')
plt.plot(x, y3, linestyle=':', color='blue', label='Dotted Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Custom Line Styles')
plt.legend()
plt.grid(True)
plt.show()
```

### More Customizations of line plot

```python
plt.plot(x, y, linestyle='-', color='r', marker='o', markerfacecolor='purple', markeredgewidth=2, label='Solid line')
plt.plot(x, y1, linestyle='--', color='g', marker='^', markerfacecolor='brown', markeredgewidth=1, label='Dashed Line')
plt.plot(x, y2, linestyle='-.', color='b', marker='.', markeredgewidth=2, label='Dash-dot Line')
plt.plot(x, y3, linestyle=':', color='y', marker='>', markerfacecolor='pink', markeredgewidth=2, label='Dotted Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Custom Line Styles')
plt.legend()
plt.show()
```

# Part 2: Scatter Plot

### Creating a Basic Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Scatter Plot with Customizations

```python
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, edgecolors = 'black', linewidth=1)
plt.title("Customized Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Scatter Plot with Annotations

```python
x_ = [1, 2, 3, 4, 5]
y_ = [2, 4, 6, 8, 10]
plt.scatter(x, y)
for i, (xi, yi) in enumerate(zip(x_, y_)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')
plt.title('Scatter Plot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Plotting Multiple Scatter Plots on the Same Graph.

```python
x2 = np.random.rand(50)
y2 = np.random.rand(50)
plt.scatter(x, y, c='r', label='Group 1')
plt.scatter(x2, y2, c='b', label='Group 2')
plt.title("Multiple Scatter Plots")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Adding Labels and Titles to the Scatter Plot

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.title("Scatter Plot with Labels")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.text(0.5, 0.5, 'Center', fontsize=12, ha='center', va='center', bbox=dict(facecolor='red', alpha=0.5))
plt.show()
```

## Part 3: Bar Chart
### Creating a Vertical Bar Chart

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]
plt.bar(categories, values)
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Creating a Horizontal Bar Chart

```python
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

### Customizing Bar Colors and Adding Edge Colors

```python
plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'], align = 'center',linewidth=2, hatch='+', edgecolor='black')
plt.title("Customized Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Part 4: Pie Chart
### Creating a Basic Pie Chart

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]
plt.pie(sizes, labels=labels)
plt.title("Basic Pie Chart")
plt.show()
```

### Customizing Slice Colors and Adding a Legend

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors)
plt.title("Pie Chart with Custom Colors")
plt.legend(labels, loc="best")
plt.show()
```

### Exploding a Slice of the Pie Chart

```python
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Pie Chart with Exploded Slice")
plt.show()
```

### Adding Percentage Labels to Slices

```python
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart with Percentage Labels")
plt.show()
```

### Adding Border Shadows and Custom Text to Slices

```python
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True, textprops={'fontsize': 10, 'color': 'red', 'style': 'oblique'})
plt.title("Pie Chart with Percentage Labels")
plt.show()
```

## Some Other Basic Plots
### Stem Plot

```python
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]
plt.stem(x, y)
plt.title('Stem Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Fill_Between Plot

```python
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))
plt.fill_between(x, y1, y2, alpha=.5, color = 'b', linewidth=0)
plt.title('Fill_Between Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Stack Plot

```python
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])
plt.stackplot(x, y)
plt.title('Stack Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Stair Plot

```python
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]
plt.stairs(y, color = 'green', linewidth=2.5)
plt.title('Stairs Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Data Visualization on a Dataset from KAGGLE

```python
df = pd.read_csv('netflix1.csv')
print(df.head())
```

### Data Preprocessing

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
```

### 1. Basic Plots

### Simple line plot

```python
plt.figure(figsize=(12, 6))
df['type'].value_counts().plot(kind='line', marker='o')
plt.title('Content Type Over Time')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

### Multiple lines on the same graph with custom styles

```python
plt.figure(figsize=(12, 6))
content_by_year = df.groupby('year_added')['type'].value_counts().unstack()
content_by_year['Movie'].plot(kind='line', marker='o', linestyle='-', color='blue', label='Movie')
content_by_year['TV Show'].plot(kind='line', marker='s', linestyle='--', color='red', label='TV Show')
plt.title('Content Type Distribution Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Content Type')
plt.show()
```

### 2. Scatter Plots

### Basic scatter plot

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['release_year'], df['duration'])
plt.title('Content Duration vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.show()
```

### Customized scatter plot

```python
plt.figure(figsize=(12, 6))
scatter = plt.scatter(df['release_year'], df['duration'], 
                      c=df['type'].map({'Movie': 'blue', 'TV Show': 'red'}),
                      s=df['rating'].map({'TV-MA': 100, 'TV-14': 80, 'TV-PG': 60, 'R': 40, 'PG-13': 20}),
                      alpha=0.6)
plt.title('Content Duration vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.colorbar(scatter)
plt.show()
```

### 3. Bar Charts

### Vertical bar chart

```python
plt.figure(figsize=(12, 6))
df['rating'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Content Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

### Horizontal bar chart with custom colors

```python
plt.figure(figsize=(12, 8))
colors = plt.cm.Spectral(np.linspace(0, 1, 10))
df['country'].value_counts().head(10).plot(kind='barh', color=colors, edgecolor='black')
plt.title('Top 10 Countries by Content Production')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

### 4. Pie Charts

### Basic pie chart

```python
plt.figure(figsize=(10, 10))
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title('Distribution of Content Type')
plt.ylabel('')
plt.show()
```

### Customized pie chart with exploded slice

```python
plt.figure(figsize=(10, 10))
types = df['type'].value_counts()
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0)  # explode first slice
plt.pie(types, labels=types.index, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow=True)
plt.title('Distribution of Content Type')
plt.legend(title='Content Types')
plt.show()
```
