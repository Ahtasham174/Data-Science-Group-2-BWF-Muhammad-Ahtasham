# Task 17: Advanced plotting techniques (subplots, histograms, box plots)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Subplots

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(x, y1, label='Sine')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].legend()

axs[1].plot(x, y2, label='Cosine', color='orange')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].legend()
plt.tight_layout()
plt.show()
```

### Customizing subplots

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(x, y1, label='Sine', color='purple', marker='>', linestyle='--', linewidth=2)
axs[0].set_title('Sine Wave', fontsize=14)
axs[0].set_xlabel('x', fontsize=12)
axs[0].set_ylabel('sin(x)', fontsize=12)
axs[0].legend()
axs[0].grid(True)

axs[1].plot(x, y2, label='Cosine', color='green', linestyle='-.', linewidth=2)
axs[1].set_title('Cosine Wave', fontsize=14)
axs[1].set_xlabel('x', fontsize=12)
axs[1].set_ylabel('cos(x)', fontsize=12)
axs[1].legend()
axs[1].grid(True)
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.random.randn(100)
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 8]
data = np.random.randn(1000)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x, y1, label='Sine Wave', color='blue')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('sin(x)')
axs[0, 0].legend()

# Scatter plot
axs[0, 1].scatter(x, y2, color='red', label='Random Scatter')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('Random Value')
axs[0, 1].legend()

# Bar plot
axs[1, 0].bar(categories, values, color='green')
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Value')

# Histogram
axs[1, 1].hist(data, bins=30, color='purple', edgecolor='black')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## 3. Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot of Random Data')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.grid(True)
plt.xticks([1, 2, 3], ['Sample 1', 'Sample 2', 'Sample 3'])
plt.show()
```

Generate a 2x2 subplot figure where:

1. The first subplot contains a line plot of y = e^-0.1x cos(2x).

2. The second subplot contains a scatter plot of y = sin(x) with random noise.

3. The third subplot contains a bar plot of a categorical dataset.

4. The fourth subplot contains a histogram of a normally distributed dataset with mean 3 and
standard deviation 1.

Customize each plot with appropriate titles, axis labels, legends, and distinct color schemes.

```python
x = np.linspace(0, 10, 100)
y1 = np.exp(-0.1 * x) * np.cos(2 * np.pi * x)
y2 = np.sin(x) + np.random.normal(0, 0.1, 100)
categories = ['A', 'B', 'C', 'D']
values = [5, 10, 7, 8]
data = np.random.normal(3, 1, 1000)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x, y1, label='Damped Cosine', color='blue')
axs[0, 0].set_title('Line Plot: Damped Cosine')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].legend()

# Scatter plot
axs[0, 1].scatter(x, y2, color='red', label='Noisy Sine')
axs[0, 1].set_title('Scatter Plot: Noisy Sine')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].legend()

# Bar plot
axs[1, 0].bar(categories, values, color='green')
axs[1, 0].set_title('Bar Plot: Categorical Data')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Value')

# Histogram
axs[1, 1].hist(data, bins=30, color='purple', edgecolor='black')
axs[1, 1].set_title('Histogram: Normally Distributed Data')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# Some Other Graph Types:

## Error Bar

```python
x = [2, 4, 6]
y = [3.6, 5, 4.2]
yerr = [0.9, 1.2, 0.5]
fig, ax = plt.subplots()
ax.errorbar(x, y, yerr, fmt='o', linewidth=2, capsize=6, color = 'r')
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
ax.set_title('Error Bar')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

## Voilin Plot

```python
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))
fig, ax = plt.subplots()
vp = ax.violinplot(D, [2, 4, 6], widths=2,
                   showmeans=False, showmedians=False, showextrema=False)
for body in vp['bodies']:
    body.set_alpha(0.9)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
ax.set_title('Voilin Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

## Event Plot

```python
x = [2, 4, 6]
D = np.random.gamma(4, size=(3, 50))
fig, ax = plt.subplots()
ax.eventplot(D, orientation="vertical", lineoffsets=x, color = 'gold', linewidth=0.75)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
ax.set_title('Event Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

## 3D Scatter Plot

```python
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs, color = 'red')
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

## 3D Surface Plot

```python
from matplotlib import cm
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Greens)
ax.set_zlabel('Z-axis', labelpad=10)
plt.tight_layout()
ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

### Tri-Surf Plot

```python
n_radii = 8
n_angles = 36
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Reds)
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])
ax.set_title('Tri-Surf Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

# Mini Project: Employee Data Analysis

### Create a Custom Dataset Using AI

```python
# Seed for reproducibility
np.random.seed(42)
n = 100  # Number of employees
employee_ids = np.arange(1, n+1)
names = [f'Employee {i}' for i in range(1, n+1)]
ages = np.random.randint(22, 65, size=n)
genders = np.random.choice(['Male', 'Female'], size=n)
departments = np.random.choice(['HR', 'Finance', 'Engineering', 'Marketing', 'Sales'], size=n)
salaries = np.random.randint(30000, 120000, size=n)
years_of_experience = np.random.randint(0, 40, size=n)
performance_scores = np.random.randint(1, 101, size=n)
hire_dates = pd.date_range(start='2000-01-01', end='2023-01-01', periods=n)

# Create DataFrame
employee_data = pd.DataFrame({
    'employee_id': employee_ids,
    'name': names,
    'age': ages,
    'gender': genders,
    'department': departments,
    'salary': salaries,
    'years_of_experience': years_of_experience,
    'performance_score': performance_scores,
    'hire_date': hire_dates
})
print(employee_data.head())
employee_data.to_csv('employee_data.csv', index=False)
```

### Loading Data

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('employee_data.csv')

# Display the first 5 rows of the dataframe to ensure it's loaded correctly
print(df.head())
```

### 1. Bar plot: Department distribution

```python
department_counts = df['department'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(department_counts.index, department_counts.values, color='green')
plt.title('Bar Plot: Department Distribution')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

### 2. Line Plot: Salary Over Time

```python
df['hire_date'] = pd.to_datetime(df['hire_date'])
df_sorted_by_date = df.sort_values('hire_date')
plt.figure(figsize=(10, 6))
plt.plot(df_sorted_by_date['hire_date'], df_sorted_by_date['salary'], label='Salary', color='blue')
plt.title('Line Plot: Salary Over Time')
plt.xlabel('Hire Date')
plt.ylabel('Salary')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

### 3. Scatter plot: Salary vs Years of Experience

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['years_of_experience'], df['salary'], color='red', label='Salary')
plt.title('Scatter Plot: Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

### 4. Histogram: Age distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'].dropna(), bins=30, color='purple', edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

### 5. Box plots: Salary distribution for different departments

```python
plt.figure(figsize=(10, 6))
department_unique = df['department'].unique()
department_data = [df[df['department'] == dept]['salary'] for dept in department_unique]
boxprops = dict(linestyle='-', linewidth=2, color='darkgoldenrod')
medianprops = dict(linestyle='-', linewidth=2.5, color='firebrick')
plt.boxplot(department_data, vert=True, patch_artist=True, boxprops=boxprops, medianprops=medianprops)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xticks(range(1, len(department_unique) + 1), department_unique, fontsize=12)
plt.show()
```

### 6. Pie Chart: Gender Distribution

```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Pie Chart: Gender Distribution')
plt.show()
```

```python
ages = df['age'].dropna()
gender_counts = df['gender'].value_counts()
race_counts = df['race'].value_counts()
armed_counts = df['armed'].value_counts()
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Line plot: Age over time
df_sorted_by_date = df.sort_values('date')
axs[0, 0].plot(df_sorted_by_date['date'], df_sorted_by_date['age'], label='Age', color='blue')
axs[0, 0].set_title('Line Plot: Age Over Time')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Age')
axs[0, 0].legend()

# Scatter plot: Age vs Armed
armed_encoded = df['armed'].astype('category').cat.codes
axs[0, 1].scatter(df['age'], armed_encoded, color='red', label='Armed')
axs[0, 1].set_title('Scatter Plot: Age vs Armed')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Armed (Encoded)')
axs[0, 1].legend()

# Bar plot: Race distribution
axs[1, 0].bar(race_counts.index, race_counts.values, color='green')
axs[1, 0].set_title('Bar Plot: Race Distribution')
axs[1, 0].set_xlabel('Race')
axs[1, 0].set_ylabel('Count')

# Histogram: Age distribution
axs[1, 1].hist(ages, bins=30, color='purple', edgecolor='black')
axs[1, 1].set_title('Histogram: Age Distribution')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
