# Task 18 ->Introduction to Seaborn (visualization with Seaborn)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

### Load the dataset

```python
file_path = 'HRDataset_v14.csv'
df = pd.read_csv(file_path)
```

### Display the first few rows of the dataset

```python
df.head()
```

# A. Scatter Plot
## Using Simple Scatter Plot with different scatter point sizes:

### 1. Scatter Plot of Salary vs. Engagement Survey:

```python
plt.figure(figsize=(10, 6))
sns.relplot(data=df, x='Salary', y='EngagementSurvey',kind='scatter', hue='GenderID', palette='dark',size='Absences')
plt.title('Salary vs. Engagement Survey by Gender')
plt.xlabel('Salary')
plt.ylabel('Engagement Survey Score')
plt.show()
```

## Using Relational Plot as Scatter Plot showing relation with other features according to relation between them:

### 2. Scatter Plot of Salary vs. Performance Score:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Salary', y='PerformanceScore', hue='MaritalStatusID', palette='viridis')
plt.title('Salary vs. Performance Score by Marital Status')
plt.xlabel('Salary')
plt.ylabel('Performance Score')
plt.show()
```

## More Visualization in Scatter Plot on Dataset:

### 3. Scatter Plot of Engagement Survey vs. Emp Satisfaction by Gender:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='EngagementSurvey', y='EmpSatisfaction', hue='Sex', style='Sex', palette='husl')
plt.title('Engagement Survey vs. Employee Satisfaction by Gender')
plt.xlabel('Engagement Survey')
plt.ylabel('Employee Satisfaction')
plt.show()
```

### 4. Scatter Plot of Days Late in Last 30 Days vs. Absences by Department:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='DaysLateLast30', y='Absences', hue='Department', palette='pastel')
plt.title('Days Late in Last 30 Days vs. Absences by Department')
plt.xlabel('Days Late in Last 30 Days')
plt.ylabel('Absences')
plt.show()
```

### 5. Scatter Plot of Salary vs. Age by Department:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='EngagementSurvey', y='Salary', hue='Department', palette='coolwarm')
plt.title('Salary vs. EngagementSurvey by Department')
plt.xlabel('EngagementSurvey')
plt.ylabel('Salary')
plt.show()
```

# B. Line Plot
## Simple Line Plot With Error Bar
### 1. Line Plot of Salary over Age:

```python
plt.figure(figsize=(10, 6))
df['Age'] = pd.to_datetime('today').year - pd.to_datetime(df['DOB']).dt.year
sns.lineplot(data=df, x='Age', y='Salary', hue='GenderID', palette='dark')
plt.title('Salary over Age by Gender')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

## Line Subplot with Added Markers, No Errorbar and Dashed Lines

### 2. Line Plot of Salary over Years of Service by Gender:

```python
plt.figure(figsize=(10, 6))
df['YearsOfService'] = (pd.to_datetime('today') - pd.to_datetime(df['DateofHire'])).dt.days // 365
sns.relplot(data=df, x='YearsOfService', y='Salary', kind='line', style='Sex', hue='Sex', palette='Set2',markers=True, dashes=True, errorbar=None)
plt.title('Salary over Years of Service by Gender')
plt.xlabel('Years of Service')
plt.ylabel('Salary')
plt.show()
```

# Box Plot
## Simple Relational Plot Using Catplot as similar to Relplot in line & scatter
### 1. Box Plot of Salary by Department:

```python
sns.catplot(data=df, x='Department', kind='box', y='Salary', palette='Blues')
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

## Removing Outliers

```python
sns.boxplot(data=df, x='Department', y='Salary', palette='Blues', sym='')
```

## Changing the whiskers
The whiskers are the two lines outside the box, that go from the minimum to the lower quartile (the start of the box) and then from the upper quartile (the end of the box) to the maximum

```python
sns.boxplot(data=df, x='Department', y='Salary', palette='Blues', sym='', whis=[0, 100], whiskerprops={'color': 'red'})
```
