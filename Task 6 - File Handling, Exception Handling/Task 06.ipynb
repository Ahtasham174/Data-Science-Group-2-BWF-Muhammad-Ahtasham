## Exception Handling Practice:

### Using try/except to check zero by division

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

### Using try and multiple except to check zero by division

```python
try:
    number = int(input("Enter a number: "))
    result = 10 / number
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter a valid number.")
```

### Using try/except/else/finally to check zero by division

```python
try:
    number = int(input("Enter a number: "))
    result = 10 / number
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter a valid number.")
else:
    print(f"The result is {result}")
finally:
    print("Execution completed.")
```

### Using raise to check for positive number

```python
try:
    num = int(input("Enter a positive number: "))
    if num <= 0:
        raise ValueError("The number must be positive.")
    else:
        print(num)
except ValueError as e:
    print(f"Error: {e}")
```

# Mini Project: Library Management System 

## Overview

The system uses a text file `books_record.txt` to store book records. Each record consists of a `Book_ID`, `Title`, `Author`, and `Year of Publish`, separated by commas. The main components of the system are:

- **Loading Books**: Load books from the text file into a list.
- **Saving Books**: Save the current list of books back to the text file.
- **Adding a Book**: Add a new book to the list and save it.
- **Searching for a Book**: Search for a book by `Book_ID`.
- **Deleting a Book**: Delete a book from the list by `Book_ID`.
- **Displaying All Books**: Display all the books in the list.
- **User Menu**: A menu to navigate through the above functionalities.

### Load Books

The `load_books()` function reads the `books_record.txt` file and loads its content into a list of dictionaries. Each dictionary represents a book with keys: `book_id`, `title`, `author`, and `year_pub`.

### Save Books

The `save_books(books)` function takes a list of books and writes it to the `books_record.txt` file, each book on a new line in the format: `Book_ID, Title, Author, Year_of_Publish`.

### Add a Book

The `add_books(books)` function allows users to add a new book to the collection. It ensures the `Book_ID` and `Year of Publish` are numeric and checks if the `Book_ID` is unique. If valid, it adds the book to the list and saves the updated list.

### Search for a Book

The `search_books(books)` function allows users to search for a book by entering its `Book_ID`. If the book is found, its details are printed; otherwise, a "Book not found" message is shown.

### Delete a Book

The `delete_books(books)` function lets users delete a book by its `Book_ID`. If the book exists, it is removed from the list, and the updated list is saved. If the `Book_ID` does not exist, a "Book not found" message is displayed.

### Display All Books

The `display_books(books)` function displays all books currently in the list. If no books are available, it prints "No books available."

### User Menu

The `menu()` function provides an interactive menu for users to choose different actions: add a book, search for a book, delete a book, display all books, or exit the system. It repeatedly shows the menu until the user chooses to exit.

## Execution

To run the system, the `menu()` function is called. This function handles loading books, displaying the menu, and executing user choices.

The system starts by loading existing books and allows the user to interact through a menu-based interface to manage the book records effectively.


```python
import os

file_t = "books_record.txt"

def load_books():
    books = []
    if os.path.exists(file_t):
        with open(file_t, "r") as f:
            for line in f:
                book_id, title, author, year_pub = line.strip().split(", ")
                books.append({
                    "book_id": book_id, "title": title, "author": author, "year_pub": year_pub
                })
    return books

def save_books(books):
    with open(file_t, "w") as f:
        for book in books:
            f.write(f'{book["book_id"]}, {book["title"]}, {book["author"]}, {book["year_pub"]}\n')

def add_books(books):
    try:
        book_id = input("Enter unique Book_ID (numeric): ")
        if not book_id.isnumeric():
            raise ValueError("Book ID must be numeric.")
        
        for book in books:
            if book["book_id"] == book_id:
                raise ValueError("Book ID already exists.")
        
        title = input("Enter book's title: ")
        author = input("Enter book's author name: ")
        year_pub = input("Enter book's year of publish (numeric): ")
        if not year_pub.isnumeric():
            raise ValueError("Year of publish must be numeric.")

        new_book = {
            "book_id": book_id, "title": title, "author": author, "year_pub": year_pub
        }
        books.append(new_book)
        save_books(books)
        print("Book added successfully.")
    except ValueError as e:
        print(f"Error: {e}")

def search_books(books):
    book_id = input("Enter book_id to search: ")
    for book in books:
        if book["book_id"] == book_id:
            print(f'{book["book_id"]}, {book["title"]}, {book["author"]}, {book["year_pub"]}')
            return
    print("Book not found!")

def delete_books(books):
    book_id = input("Enter book_id to delete: ")
    for book in books:
        if book["book_id"] == book_id:
            books.remove(book)
            save_books(books)
            print("Book deleted successfully.")
            return
    print("Book not found!")

def display_books(books):
    if books:
        for book in books:
            print(f'{book["book_id"]}, {book["title"]}, {book["author"]}, {book["year_pub"]}')
    else:
        print("No books available.")
        
def menu():
    books = load_books()
    
    while True:
        print("\nLibrary Management System")
        print("1. Add a book")
        print("2. Search for a book")
        print("3. Delete a book")
        print("4. Display all books")
        print("5. Exit")
        
        try:
            choice = int(input("Enter your choice: "))
            if choice == 1:
                add_books(books)
            elif choice == 2:
                search_books(books)
            elif choice == 3:
                delete_books(books)
            elif choice == 4:
                display_books(books)
            elif choice == 5:
                break
            else:
                print("Invalid choice. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    menu()
```
